```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_state(initial_amplitudes, new_state_amplitudes):
    # We are only consiering the change in theta, not the global phase shift
    
    # First the amplitude is converted from caresian to polar form
    imaginary_part = np.imag(initial_amplitudes)
    real_part = np.real(initial_amplitudes)
    r1, r2 = np.sqrt(imaginary_part**2 + real_part**2)
    #print(imaginary_part)
    theta_initial = 2*(np.arccos(r1)) + np.pi/2

    imaginary_part = np.imag(new_state_amplitudes)
    real_part = np.real(new_state_amplitudes)
    r1, r2 = np.sqrt(imaginary_part**2 + real_part**2)
    theta_new = 2*(np.arccos(r1)) + np.pi/2

    theta_diff = theta_initial - theta_new
    lable_angle = theta_new + theta_diff/2
    # find numbers for arc plot
    x_arc = np.cos(np.linspace(np.minimum(theta_initial, theta_new), np.maximum(theta_initial, theta_new), 20))*0.25
    y_arc = np.sin(np.linspace(np.minimum(theta_initial, theta_new), np.maximum(theta_initial, theta_new), 20))*0.25

    fig, ax = plt.subplots(1, 1, figsize=(5, 5))
    circ = plt.Circle((0, 0), radius=1, edgecolor='black', facecolor='None')
    ax.add_patch(circ)
    ax.plot([-1.5,1.5], [0, 0], color="black")
    ax.plot([0, 0], [1.5, -1.5], color="black")
    ax.set_xlim([-1.5, 1.5])
    ax.set_ylim([-1.5, 1.5])
    plt.text(0.1, 1.1, r'$\left| 0 \right\rangle$')
    plt.text(0.1, -1.1, r'$\left| 1 \right\rangle$')
    plt.text(1.1, 0.1, r'$\left| - \right\rangle$')
    plt.text(-1.2, 0.1, r'$\left| + \right\rangle$')
    plt.text(1.4, 0.1, "$x$")
    plt.text(0.1, 1.3, "$z$")
    plt.text(np.cos(lable_angle)*0.7, np.sin(lable_angle)*0.5, r'$10\cdot\Delta\theta$')
    plt.plot([0, np.cos(theta_initial)], [0, np.sin(theta_initial)], color="r", linewidth=2)
    plt.plot([0, np.cos(theta_new)], [0, np.sin(theta_new)], color="g", linewidth=2)
    plt.plot(x_arc, y_arc, color="r", linewidth=0.5)
    plt.show()
```

```python
initial_state = np.array([1/np.sqrt(2), 1/np.sqrt(2)])
rotation_angle = 0.01*np.pi*10
new_state = initial_state @ np.array([[np.cos(rotation_angle), -np.sin(rotation_angle)],[np.sin(rotation_angle), np.cos(rotation_angle)]])
print(new_state)
plot_state(initial_state, new_state)
```

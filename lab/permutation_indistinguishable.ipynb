```python
def get_permutation(n: int, j: int):
        """This method is used to map an integer n to a permutation of j job numbers (j! possible permutations).
        Thus n must be between 0 and j!-1

        Parameters
        ----------
        n : int
            Permuation number
        j : int
            number of jobs (permutation size)
        """
        i = j
        p_next = n
        result = np.empty(i, dtype=int)
        selectables = np.arange(i)
        for k in range(len(result)):
            j = i - k - 1
            cur_select_index = int(p_next / math.factorial(j))
            cur_job_number = selectables[selectables != -1][cur_select_index]
            result[k] = cur_job_number
            # Remove cur_job number form available jobs for next iteration
            selectables[cur_job_number] = -1
            p_next = p_next % math.factorial(j)

        return result
```

```python
j = 3
m = 3

new_list = []
for i in range(j):
    for k in range(m):
        new_list.append(i)

max_num_individuals = math.factorial(j*m)#math.factorial(j*m)/(math.factorial(m)**j)
step_length = math.factorial(m)**j
print(max_num_individuals)
ref_list = np.array(new_list)
temp_list = []
n_unique_accum = []
for i in range(int(max_num_individuals)):
    cur_index = i #* step_length
    cur_result = ref_list[get_permutation(cur_index, j*m)]
    print(cur_result)
    temp_list.append(str(cur_result))
    n_unique_accum.append(len(set(temp_list)))

print(len(set(temp_list)))
print(n_unique_accum)
plt.plot(np.arange(0, int(max_num_individuals)), n_unique_accum)
```
